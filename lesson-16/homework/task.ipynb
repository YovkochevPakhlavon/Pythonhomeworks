#### Part 3: Challenges and Explorations  

- From **`iris.json`**: Calculate the mean, median, and standard deviation for each numerical column.  

```python
#First method: 
print('Mean values of each column:')
data_iris.select_dtypes('number').mean()
```

```python
print('Median values of each column:')
data_iris.select_dtypes('number').median()
```

```python
print('Standart deviation of each column:')
data_iris.select_dtypes('number').std()
```

```python
# or, We can use numpy function directly:
for i in data_iris.columns:
    if data_iris[i].dtype != 'object':
        print(f"Mean of column '{i}': {data_iris[i].mean()} ")
        print(f"Median of column '{i}': {data_iris[i].median()} ")
        print(f"Standard deviation of column '{i}': {data_iris[i].std()} ")
        print('-----------------------------------')
```

- From **`titanic.xlsx`**: Find the minimum, maximum, and sum of passenger ages.

- From **`movie.csv`**:  
    - Identify the director with the highest total `director_facebook_likes`.  
    - Find the 5 longest movies and their respective directors. 

```python
#First, we find the the highest total `director_facebook_likes` and we find this row and return with the director name:
data_movie[data_movie['director_facebook_likes'] == data_movie['director_facebook_likes'].max()][['director_name','director_facebook_likes']]
```

```python
#First, we find the the longest movies and we find the corresponding rows and return with the directors' name:
for i in list(data_movie['duration'].sort_values(ascending=False).head(5)):
    print(f"{data_movie[data_movie['duration'] == i][['director_name','duration']]}")
```
