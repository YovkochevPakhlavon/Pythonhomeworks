```python
import matplotlib.pyplot as plt 
import numpy as np
```

#### **1. Basic Plotting**
- **Task**: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

```python
x = np.arange(-10, 10, 0.001)
y = x ** 2 - 4 * x + 4
plt.plot(x,y,'r-',label='$ f(x) = x^2 - 4x + 4 $')
plt.title('Parabola')
plt.xlabel('X')
plt.ylabel('Y', rotation=0) #rotation helps to make Y like this(Y) vertical.
plt.legend() #needs for showing label of the figure
plt.show()
```

#### **2. Sine and Cosine Plot**
- **Task**: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.

```python
x = np.linspace(0,2*np.pi,1000)
plt.plot(x,np.sin(x),'r--',label='$ \sin(x) $') #graph will be red and dashed line
plt.plot(x,np.cos(x),'b-',label='$ \cos(x) $') #graph will be blue and solid line
plt.title('Trigonometirc functions')
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.legend()
plt.show()
```

#### **3. Subplots**
- **Task**: Create a 2x2 grid of subplots. In each subplot, plot:
  - Top-left: $ f(x) = x^3 $
  - Top-right: $ f(x) = \sin(x) $
  - Bottom-left: $ f(x) = e^x $
  - Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)

  Customize each plot with titles, axis labels, and different colors.


```python
plt.figure(figsize=(10, 9)) #this will determine the size of the figure.
x = np.linspace(-5,5,1000)
plt.subplot(2, 2, 1)  #Top-lef
plt.plot(x, x ** 3, 'r-',label= '$f(x) = x^3$') #red solid line
plt.title('Cubic function')
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.legend()

plt.subplot(2, 2, 2) #Top-right
plt.plot(x, np.sin(x), 'b--',label='$f(x) = \sin(x)$' ) #blue dashed
plt.title('Trigonometric function')
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.legend(loc='upper left') #we can show the location of legend.

plt.subplot(2, 2, 3) #Bottom-left
plt.plot(x, np.e ** x, 'k:', label='$f(x) = e^x$') #black dotted
plt.title('Exponential function')
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.legend()

plt.subplot(2, 2, 4) #Bottom-right
x = np.linspace(0,5,1000)
plt.plot(x, x ** 3, 'g-.' , label='$f(x) = \log(x+1)$') #green dash-dotted
plt.title('Logarithm function')
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.legend()
plt.show()
```

#### **4. Scatter Plot**
- **Task**: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.


```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
# Generate random colors and markers
colors = np.random.rand(100)
markers = ['o', 's', 'D', 'v', '^', '<', '>']

plt.figure(figsize=(8, 6))
for i in range(100):
    plt.scatter(x[i], y[i], color=plt.cm.viridis(colors[i]), marker=np.random.choice(markers))
plt.title("Scatter Plot of 100 Random Points")
plt.xlabel("X")
plt.ylabel("Y")
# Add grid
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### **5. Histogram**
- **Task**: Generate a random dataset of 1000 values sampled from a normal distribution (mean=0, std=1). Plot a histogram of the data with 30 bins. Add a title and axis labels. Adjust the transparency of the bars using the `alpha` parameter.


```python
data = np.random.normal(0,1,1000) #Generates 1000 values with mean=0, std=1
plt.hist(data,bins=30,color='skyblue', alpha=0.7, edgecolor='red')
plt.title('Histogram of random 1000 values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#### **6. 3D Plotting**
- **Task**: Create a 3D surface plot for the function $ f(x, y) = \cos(x^2 + y^2) $ over the range of $ x $ and $ y $ values from -5 to 5. Use a suitable colormap and add a colorbar. Set appropriate labels for the axes and title.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

xx, yy = np.meshgrid(x, y) # Creates a meshgrid

# Computes function values
zz = np.cos(xx**2 + yy**2)

surf = ax.plot_surface(xx, yy, zz, cmap='viridis') # Plots surface and store the mappable object

# Add a colorbar
fig.colorbar(surf, ax=ax, shrink=0.6, aspect=10)

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')
plt.show()
```

#### **7. Bar Chart**
- **Task**: Create a vertical bar chart displaying the sales data for five different products: `['Product A', 'Product B', 'Product C', 'Product D', 'Product E']`. The sales values for each product are `[200, 150, 250, 175, 225]`. Customize the chart with a title, axis labels, and different bar colors.


```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'green', 'yellow', 'magenta'] #list of colors for each bar.
plt.bar(products,values,color=colors)
# Labels and title
plt.title('Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales values')
plt.show()
```

#### **8. Stacked Bar Chart**
- **Task**: Create a stacked bar chart that shows the contribution of three different categories (`'Category A'`, `'Category B'`, and `'Category C'`) over four time periods (`'T1'`, `'T2'`, `'T3'`, `'T4'`). Use sample data for each category at each time period. Customize the chart with a title, axis labels, and a legend.

```python
plt.figure(figsize=(8,6))
#Sample data for each categories over four time periods:
A = [100,120,105,135]
B = [120,140,105,125]
C = [90,120,127,115]
#Time periods:
time = ['2000','2005','2010','2015']
x = np.arange(len(time)) #Helps to locate bar charts
plt.bar(x-0.22, A, color='b', width=0.2, label='Category A')
plt.bar(x, B, color='r', width=0.2, label='Category B')
plt.bar(x+0.22, C, color='g', width=0.2, label='Category C')
# Labels and title
plt.title('Stacked Bar Chart')
plt.xticks(x, time)
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```
